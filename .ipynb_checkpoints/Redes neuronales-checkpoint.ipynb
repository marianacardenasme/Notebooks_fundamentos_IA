### LABORATORIO – REDES NEURONALES

```python
import numpy as np
import sympy as sp
import random
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

Se crea la función que devuelve las coordenadas [x, y] en las que se encuentra el objeto a partir de los parámetros iniciales y un tiempo t.

```python
x, y, vx, vy, ax, ay, dt= sp.symbols('x y vx vy ax ay dt')

def hallar_posicion(v0, theta, m, c, t):
    g = 9.8
    dt = 0.1
    
    x, y = 0, 0
    vx = v0 * np.cos(np.deg2rad(theta))
    vy = v0 * np.sin(np.deg2rad(theta))
    
    for i in np.arange(0, t, dt):
        ax = - (c/m) * vx
        ay = - g - (c/m) * vy
        
        vx = vx + ax * dt
        vy = vy + ay * dt
        
        x = x + vx * dt
        y = y + vy * dt
    
        
            
    return float(round(x, 5)), float(round(y, 5))
```

Se genera un dataframe con 5000 registros que contienen datos aleatorios en los parámetros iniciales y su posición en un tiempo t.

```python
datos = []

for j in range(0, 5001):
    
    # Generar aleatoriamente los parámetros iniciales velocidad inicial, ángulo, masa, constante de resistencia del aire y tiempo
    v0 = round(random.uniform(1, 5), 2)
    theta = random.randint(1, 60)
    m = round(random.uniform(1, 10), 2)
    c = round(random.uniform(0.1, 0.2), 2)
    t = round(random.uniform(0.1, 3), 1)
    
    # Calcular la posición en el tiempo t teniendo en cuenta las condiciones iniciales
    nuevo_registro = hallar_posicion(v0, theta, m, c, t)
    datos.append((v0, theta, m, c, t, nuevo_registro[0], nuevo_registro[1]))

# Convertir la lista en dataframe
dataframe = pd.DataFrame(datos, columns=['velocidad inicial', 'ángulo', 'masa', 'resistencia', 'tiempo', 'coordenada x', 'coordenada y'])
dataframe
```

Todas las columnas son numéricas, se deben escalar para que no haya columnas que influyan más que otras en la predicción que arroje el modelo.

```python
X = dataframe[['velocidad inicial', 'ángulo', 'masa', 'resistencia', 'tiempo']].values
y = dataframe[['coordenada x', 'coordenada y']].values

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
```

Se separan los datos en 80% entrenamiento y 20% pruebas.

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.2, random_state = 42)
```

Ahora se crea el modelo secuencial con 2 capas intermedias con funciones de activación relu (devuelve 0 cuando el valor es negativo, de lo contrario devuelve el mismo valor que recibe) y dos neuronas de salida con función de activación linear (ya que hay salidas con valores negativos).

```python
model = Sequential()

# Capas intermedias
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))

# Capa de salida
model.add(Dense(2, activation = 'linear'))

model.compile(optimizer = 'adam', loss = 'mse')

history = model.fit(X_train, y_train, epochs = 500, batch_size = 32, validation_data = (X_test, y_test))
```

Se grafica la pérdida en entrenamiento y en validación

```python
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Pérdida entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida validación')
plt.xlabel('Épocas')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()
```

Se usa el modelo para predecir.

```python
pred_scaled = model.predict(X_test)
pred = scaler_y.inverse_transform(pred_scaled)
```

Finalmente se grafican las salidas tanto de entrenamiento coomo de prueba. Estas corresponden a las coordenadas [x, y] que tienen los proyectiles en un tiempo t a partir de sus condiciones iniciales.

```python
plt.figure(figsize=(8,6))

# Real
plt.scatter(
    dataframe['coordenada x'],
    dataframe['coordenada y'],
    label='Real',
    alpha=0.5
)

# Predicciones
plt.scatter(
    pred[:,0],
    pred[:,1],
    label='Predicho',
    alpha=0.5
)

plt.title('Coordenadas reales vs predichas')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()
plt.grid(True)
plt.show()
```
