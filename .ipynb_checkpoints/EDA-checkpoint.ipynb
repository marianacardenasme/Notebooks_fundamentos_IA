## LABORATORIO – EDA

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")
```

```python
# Cargar el dataset
data = pd.read_csv("netflix_titles_nov_2019.csv")
data.head()
```

```python
# Análisis exploratorio
data.info()
```

La columna show_id no aporta mucho valor así que se elimina.

```python
data = data.drop(columns=['show_id'])
```

De las 12 columnas que hay en el dataset hay 5 que tienen valores nulos: director, cast, country, date_added y rating. 

La columna director tiene 1901 datos nulos, lo cual equivale a un 32.6% de los datos. Estos son demasiados datos como para eliminar los registros que tengan valores nulos en esta columna. Teniendo en cuenta que lo que se desea hacer es una segmentación de películas/series y que esta columna tiene gran cantidad de valores distintos, se decide que la mejor manera de tratar este inconveniente es eliminando la columna (las otras columnas pueden aportar más a la segmentación).

Lo mismo ocurre con la columna cast, que tiene 556 valores nulos, equivalentes al 9.5% de los datos. Por lo que esta también se elimina.

```python
data2 = data.drop(columns=['director', 'cast'])
```

Los 427 valores nulos de la columna country se reemplazan por un valor genérico.

```python
data2['country'] = data2['country'].fillna('Unknown')
```

La columna date_added se reemplaza por una nueva columna llamada year_added en la cual se registra solo el año (descartando día y mes).

```python
# Convertir la columna a tipo datetime
data2['date_added'] = pd.to_datetime(data2['date_added'], errors='coerce')

# Crear una nueva columna solo con el año
data2['year_added'] = data2['date_added'].dt.year

# Eliminar la columna original
data2.drop(columns=['date_added'], inplace=True)
```

Los valores nulos de esta columna se reemplazan por la moda (el valor que más se repite).

```python
# Calcular la moda 
moda_year_added = data2['year_added'].mode()[0]

# Reemplazar valores
data2['year_added'] = data2['year_added'].fillna(moda_year_added).astype(int)
```

Los valores nulos de la columna rating también se reemplazan por la moda.

```python
# Calcular la moda 
moda_rating = data2['rating'].mode()[0]

# Reemplazar valores
data2['rating'] = data2['rating'].fillna(moda_rating)
```

Después de tratar los valores nulos, se revisan las columnas categóricas para ver de qué manera se pueden abordar. Las columnas title y description no van a ser muy útiles a la hora de realizar la segmentación, así que se eliminan.

```python
data3 = data2.drop(columns=['title', 'description'])
```

La columna country tiene registros donde hay más de un valor (pudo haber sido una película grabada en 2 o 3 países) así que se realiza un explode para separar valores múltiples en filas.

```python
# Poner en tipo string
data3['country'] = data3['country'].astype(str)

# Separar los valores
data3['country'] = data3['country'].str.split(',')

# Expandir en filas individuales
data4 = data3.explode('country')

# Limpiar espacios
data4['country'] = data4['country'].str.strip()
```

Se hace el mismo tratamiento para la columna listed_in.

```python
# Poner en tipo string
data4['listed_in'] = data4['listed_in'].astype(str)

# Separar los valores
data4['listed_in'] = data4['listed_in'].str.split(',')

# Expandir en filas individuales
data5 = data4.explode('listed_in')

# Limpiar espacios
data5['listed_in'] = data5['listed_in'].str.strip()
```

Se realiza la codificación de ls variables categóricas.

```python
LE = LabelEncoder()
data6 = data5.copy()
for i in ['country', 'rating', 'duration', 'listed_in', 'type']:
    data6[i]=data6[[i]].apply(LE.fit_transform)

print('Dataset con variables numéricas')
data6.head()
```

Se procede con la estandarización de los datos.

```python
# Crear una copia de los datos
data_copy = data6.copy()

#Escalar la data
scaler = StandardScaler()
scaler.fit(data_copy)
scaled_ds = pd.DataFrame(scaler.transform(data_copy),columns=data_copy.columns)
print("Todas las características estan estandarizadas")
```

```python
scaled_ds.head()
```

**Análisis exploratorio de datos**

Teniendo todas las columnas estandarizadas es posible sacar las correlaciones entre variables para hacer un mapa de calor:

```python
corr = scaled_ds.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
```

Como se puede observar, el valor más alto en el mapa de calor corresponde a la correlación entre las variables type y listed_in (0.33), lo que indica que las variables tienden a moverse en la misma dirección. Por otro lado, hay varios pares de variables con correlaciones negativas, como type y duration (-0.37), lo que indica que estas variables tienen una relación inversamente proporcional.

También se puede graficar un historial que muestre la distribución de películas o series por año en que fueron añadidos.

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data6, x='year_added', color='skyblue')
plt.title('Distribución de películas/series por año añadido')
plt.xlabel('Año en que fue añadido')
plt.ylabel('Cantidad de películas/series')
plt.xticks(rotation=45)
plt.show()
```

Y los 10 países con mayor cantidad de series/películas.

```python
top_countries = data5['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('10 países con más películas/series')
plt.xlabel('Cantidad de películas/series')
plt.ylabel('País')
plt.show()
```

Ahora se realiza un análisis de componentes principales (PCA).

```python
pca = PCA().fit(scaled_ds)
eigenvalues = pca.explained_variance_
componentes = np.sum(eigenvalues > 1)
print('La cantidad de componentes resultantes es: ', componentes)
```

```python
pca = PCA(n_components=componentes)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe()
```

Finalmente se aplica un modelo de clustering aglomerativo. Para esto se genera un diagrama de codo donde se halla la cantidad más óptima de clusters.

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=3, random_state=42, n_init=10)
codo = KElbowVisualizer(modelo, k=10, timings=False)
codo.fit(PCA_ds)
codo.show()
```

Como se puede observar, la cantidad óptima es 4 custers. Con esta información se procede a ejecutar el modelo de clasificación.

```python
from sklearn.cluster import AgglomerativeClustering
AC = AgglomerativeClustering(n_clusters=4)
predicciones = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = predicciones
data6["Clusters"]= predicciones
```

Ahora se grafican los 4 clusters.

```python
fig = plt.figure(figsize=(5, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(PCA_ds['col1'], PCA_ds['col2'], PCA_ds['col3'],
                     c=PCA_ds['Clusters'], cmap='viridis', edgecolor='k', s=50)

ax.set_title('Resultado del Clustering Aglomerativo')
ax.set_xlabel('CP 1')
ax.set_ylabel('CP 2')
ax.set_zlabel('CP 3')
ax.legend(*scatter.legend_elements(), title='Clusters')

plt.show()
```

Y se analizan las diferentes columnas en los clusters.

```python
analisis_cluster = data6.groupby('Clusters').describe()
for col in analisis_cluster.columns.levels[0]:
    print("Análisis de cluster para {}:".format(col))
    print(analisis_cluster[col])
    print("--------------------------------------------------------")
```

Se puede usar esta clasificación para relacionar variables como country y year_added.

```python
plt.figure(figsize=(5,3)) 
pl = sns.scatterplot(data = data6, x=data6["country"], y=data6["year_added"], hue=data6["Clusters"],
                     palette=['Purple','Blue','Green','Yellow'])
pl.set_title("Perfilamineto basado en país y año en que se añadió")
plt.legend()
plt.show()
```
