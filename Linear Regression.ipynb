```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### 1. Ventas mensuales de una tienda durante un año

```python
x1 = np.array(list(range(1, 13)))
y1 = np.array(list(range(20, 76, 5)))
```

```python
# Modelo
b1_1, b0_1 = np.polyfit(x1, y1, 1)
plt.scatter(x1, y1, color='plum', label="Datos")
plt.plot(x1, b1_1 * x1 + b0_1, color='pink', label="Regresion")
plt.legend()
```

```python
# Correlación
correlacion1 = np.corrcoef(x1, y1)[0, 1]
print("Correlación de Pearson:", correlacion1)
```

La correlación obtenida es 1, lo que indica que hay una relación lineal positiva perfecta entre las variables, en este caso ambas aumentan de forma proporcional.

```python
# Estadística
media_x1 = np.mean(x1)
media_y1 = np.mean(y1)
std1 = np.std(x1)
var1 = np.var(x1)
cov1 = np.cov(x1,y1)
print("Estadística básica: \nMedia x: {} \nMedia y: {} \nStd: {} \nVar: {} \nCov: {}".format(media_x1, media_y1, std1, var1, cov1))
```

La varianza, al ser alta, indica que los valores están esparcidos en un rango amplio de valores, lo cual tiene sentido ya que los valores no se acumulan cerca de la media, sino que están igualmente espaciados unos de otros.

```python
# Coeficientes
print("Coeficientes del modelo: \nPendiente(b1): {} \nIntercepto(b0): {}".format(b1_1, b0_1))
```

```python
# Precisión
x1_lr = np.array(list(range(1, 13))).reshape(-1,1)
modelo = LinearRegression()
modelo.fit(x1_lr, y1)
R2_1 = modelo.score(x1_lr, y1)
precision1 = R2_1 * 100
print("Porcentaje de precisión: ", precision1)
```

```python
# Predicción con datos de prueba
def predict_y(x):
    return b0_1 + b1_1 * x

new_x1 = np.array([13, 14, 15, 16, 17, 18])
new_y1 = predict_y(new_x1)
print("Predicción de y: ", new_y1)

b1_1prueba, b0_1prueba = np.polyfit(new_x1, new_y1, 1)
plt.scatter(new_x1, new_y1, color='plum', label="Datos")
plt.plot(new_x1, b1_1prueba * new_x1 + b0_1prueba, color='pink', label="Regresion")
plt.legend()
```
